```python
import neurodynex3.ojas_rule.oja as oja
import matplotlib.pyplot as plt
import numpy as np


cloud = oja.make_cloud(n=200, ratio=.3, angle=60)
wcourse = oja.learn(cloud, initial_angle=-20, eta=0.04)
plt.scatter(cloud[:, 0], cloud[:, 1], marker=".", alpha=.2)
plt.plot(wcourse[-1, 0], wcourse[-1, 1], "or", markersize=10)
plt.axis('equal')
plt.figure()
plt.plot(wcourse[:, 0], "g")
plt.plot(wcourse[:, 1], "b")
plt.plot((wcourse[:, 0]**2 + wcourse[:, 1]**2)**0.5, "r")
print("The final weight vector w is: ({},{})".format(wcourse[-1,0],wcourse[-1,1]))
```

```python
cloud  = oja.make_cloud(n=400, ratio=1)
wcourses = []
for _ in range(10):
    wcourses.append(oja.learn(cloud, eta=0.04))


plt.scatter(cloud[:, 0], cloud[:, 1], marker=".", alpha=.2)
for i in range(10):
    plt.plot(wcourses[i][-1, 0], wcourses[i][-1, 1], "or", markersize=10)
    plt.axis('equal')

plt.figure()
for i in range(10):
    plt.plot(wcourses[i][:, 0], "g", linewidth=0.5, alpha=0.4)
    plt.plot(wcourses[i][:, 1], "b", linewidth=0.5, alpha=0.4)
    plt.plot((wcourses[i][:, 0]**2 + wcourses[i][:, 1]**2)**0.5, "r", linewidth=0.5, alpha=0.4)
```
