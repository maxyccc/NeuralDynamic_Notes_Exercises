```python
%matplotlib inline
import brian2 as b2
import matplotlib.pyplot as plt
import numpy as np
from neurodynex3.leaky_integrate_and_fire import LIF
from neurodynex3.tools import input_factory, plot_tools

# LIF.getting_started()
LIF.print_default_parameters()
mini_current = (LIF.FIRING_THRESHOLD - LIF.V_REST) / LIF.MEMBRANE_RESISTANCE
step_current = input_factory.get_step_current(
    t_start=5, t_end=100, unit_time=b2.ms, amplitude=mini_current
)
(state_monitor,spike_monitor) = LIF.simulate_LIF_neuron(
    input_current=step_current, simulation_time = 100 * b2.ms
)
plot_tools.plot_voltage_and_current_traces(
    state_monitor, step_current, title='min input',
    firing_threshold=LIF.FIRING_THRESHOLD,
)
print(
    "minimal current: {} \nnr of spikes: {}"
    .format(mini_current.in_best_unit(precision=3),
            spike_monitor.count[0])
)
```

```python
def plot_freq(frequency):
    sin_input = input_factory.get_sinusoidal_current(
        200, 1000, unit_time=0.1*b2.ms, amplitude=2.5*b2.namp,
        frequency=frequency*b2.Hz, direct_current=0. * b2.namp
    )
    (state_monitor, spike_monitor) = LIF.simulate_LIF_neuron(input_current=sin_input, simulation_time = 120 * b2.ms, firing_threshold=0*b2.mV)
    plot_tools.plot_voltage_and_current_traces(
        state_monitor, sin_input, title="Sinusoidal input current"
    )
    print("nr of spikes: {}".format(spike_monitor.count[0]))
```

```python
plot_freq(1000)
plot_freq(200)
plot_freq(100)
plot_freq(50)
plot_freq(10)

```
