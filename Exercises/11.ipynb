```python
import numpy as np


def get_orientation(idx_list, N):
    return [np.round((id_+0.5) * (360/N), 2) for id_ in idx_list]
print(get_orientation([0, 1, 5, 10], 11))
print(get_orientation([0, 1, 499, 500, 999], 1000))
```

```python
def get_spike_count(spike_monitor, spike_index_list, t_min, t_max):
    nr_neurons = len(spike_index_list)
    spike_count_list = np.zeros(nr_neurons)
    spike_trains = spike_monitor.spike_trains()
    for i in spike_index_list:
        spike_count_list[i] = sum((spike_trains[i] >= t_min) & (spike_trains[i] < t_max))
    return spike_count_list

t = 400*b2.ms  # time point of interest
t_window = 30*b2.ms # width of the window over which the average is taken

rate_monitor_excit, spike_monitor_excit,\
voltage_monitor_excit, idx_monitored_neurons_excit,\
rate_monitor_inhib, spike_monitor_inhib, \
voltage_monitor_inhib, idx_monitored_neurons_inhib,\
w_profile = wm_model.simulate_wm(
    stimulus_center_deg=120, stimulus_width_deg=30,
    stimulus_strength=.3 * b2.namp, t_stimulus_start=100 * b2.ms,
    t_stimulus_duration=200 * b2.ms, sim_time=500. * b2.ms)

fig, ax_raster, ax_rate, ax_voltage = plot_tools.plot_network_activity(rate_monitor_excit, spike_monitor_excit, voltage_monitor_excit, t_min=0. * b2.ms)
plt.show()

t_min = t-t_window/2
t_max = t+t_window/2
spike_counts = get_spike_count(spike_monitor_excit,
                               idx_monitored_neurons_excit,
                               t_min, t_max)
spike_rates = spike_counts/(t_max-t_min)/b2.second
plt.plot(spike_rates, ".b")
plt.title("Bump profile in the time interval[{},{}]".format(t_min, t_max))
plt.xlabel("Neuron index")
plt.ylabel("Spike rate [Hz]")
```

```python
def get_theta_time_series(spike_monitor, idx_monitored_neurons,
                          t_snapshots, t_window):
    theta_ts = []
    for t_snapshot in t_snapshots:
        spike_counts = get_spike_count(spike_monitor, idx_monitored_neurons,
                                       t_snapshot-t_window/2, t_snapshot+t_window/2)
        orient = get_orientation(idx_monitored_neurons, len(idx_monitored_neurons))
        theta = sum([spike_counts[i] * orient[i] for i in range(len(spike_counts))])/sum(spike_counts)
        theta_ts.append(theta)
    return theta_ts

import math
# Example how to create an array of timestamps spaced by snapshot_interval in the interval of interest.
t_snapshots = range(
    int(math.floor((100 * b2.ms+200 * b2.ms)/b2.ms)),  # lower bound
    int(math.floor((500. * b2.ms- 30*b2.ms/2)/b2.ms)),  # Subtract half window. Avoids an out-of-bound error later.
    int(round(10))  # spacing between time stamps
    )*b2.ms

# how your function get_theta_time_series could be called:
theta_ts = get_theta_time_series(spike_monitor_excit,
                                 idx_monitored_neurons_excit,
                                 t_snapshots, t_window)

plt.plot(t_snapshots/b2.ms, theta_ts)
```
